# Análise dos Faturamentos de 3 Lojas/Restaurantes em Junho/20

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# variável com dataset do faturamento das lojas/restaurantes
f_lojas = pd.read_excel('C:/Users/Karla/Documents/Marcio/Datasets/faturamanto_restaurantes.xls')
```

```python
# visualização do dataset
f_lojas
```

```python
# conversão para formato data e hora
f_lojas['data'] = pd.to_datetime(f_lojas['data'])
```

```python
# tornar a coluna data como índice
f_lojas.set_index(['data'],inplace=True)
```

```python
# visualização do faturamento das 3 lojas em junho/2020
f_lojas.plot(figsize=(10,5), title='Faturamentos das Lojas em Junho/20', marker='o')
```

```python
# visualização dos faturamentos de 01/06/20 a 07/06/20
f_lojas["2020-06-01":"2020-06-07"].plot(figsize=(10,5), title='Faturamentos das Lojas de 01/06/20 a 07/06/20', marker='o')
```

```python
# visualização da loja/restaurante do Meier
f_lojas['fatur_lojaMeier'].plot(figsize=(10,5), title='Faturamento da Loja Méier em Junho/20', marker='o')
```

```python
# visualização dos percentis de faturamento por boxplot das 3 lojas
f_lojas.plot.box(figsize=(10,5), title='Faturamento das Lojas visto por boxplots')
```

```python
# visualização dos faturamentos das 3 lojas ao longo de junho/20
axs = f_lojas.plot(figsize=(18,6), subplots=True, title='Faturamento das Lojas ao Longo de Junho/20')
```
